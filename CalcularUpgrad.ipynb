```python
import math
```

```python
def add(a,b):
    sum1 = a+b
    return sum1

def subtract(a,b):
    diff = a-b
    return diff

def multiply(a,b):
    product = a*b
    return product

def divide(a,b):
    q = a/b
    return q

def hcf(a,b):
    if a == 0:
        return b
    return hcf(b % a, a)

def lcm(a,b):
    return (a * b)/ hcf(a,b)

def power(a,b):
    return a**b

def log(a):
    return math.log(a)
```

```python
a = int(input("Enter first number: "))
while(1>0):
    print("""
    1 for Addition 
    2 for Subtraction 
    3 for Multiplication 
    4 for Division 
    5 for power
    6 for Highest Common Factor 
    7 for Lowest Common Division 
    8 for Log 
    9 for exit
    """)
    x = int(input('Enter Operator: '))
    print()
    if(x < 8):
        b = int(input("Enter second number: "))
    if(x == 1):
        ans = add(a,b)
        print("Your Required Result: ", ans)
    elif(x == 2):
        ans = subtract(a,b)
        print("Your Required Result: ",ans)
    elif(x==3):
        ans = multiply(a,b)
        print("Your Required Result: ",ans)
    elif(x == 4):
        ans = divide(a,b)
        print("Your Required Result: ",ans)
    elif(x==5):
        ans = power(a,b)
        print("Your Required Result: ",ans)
    elif(x==6):
        ans = hcf(a,b)
        print("Your Required Result: ",ans)
    elif(x==7):
        ans = lcm(a,b)
        print("Your Required Result: ",ans)
    elif(x==8):
        ans = log(a)
        print("Your Required Result: ",ans)
    elif(x == 9):
        print("Yours Result: ",a)
        break;
    else:
        print("Invalid Operator")
    a = ans
```
