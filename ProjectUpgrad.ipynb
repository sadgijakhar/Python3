# Assignment: Problem Statement
Let’s put ourselves in the shoes of a childcare provider.
As childcare providers, we have an essential role in children’s lives by helping them learn and grow. You help children get the nourishment and energy they need to learn and grow by providing healthy meals. 


Also, keep in mind that preschoolers who eat a variety of healthy foods and play actively several times every day are less likely to be overweight or obese.

In this assignment, you will try to calculate a child's nutritions, using Python. This is a compulsory individual assignment wherein you will write Python code to take input from the user,  set various conditions, calculate BMI and calorie consumption, and submit the code.


Hints:
Create a child nutrition calculator based on the following tasks & conditions:
The calculator should be able to store the details of the child such as the name, age, gender, height and weight.
Calculate the BMI of the child using the standard BMI formula.
Display the minimum daily calorie requirement based on age. Use the below criteria to display the same:

 

Age b/w 0-2 years: 800 calories
Age b/w 2-4 years: 1400 calories
Age b/w 4-8 years: 1800 calories
Take as input what that child has eaten throughout the day and in what quantities.

 

Calculate the daily calorie consumption based on the following criteria:

Milk: 100 cal/ 100g
Egg: 155 cal/ 100g
Rice: 130 cal/ 100g
Lentils: 113 cal/ 100g
Vegetable: 85 cal/100g
Meat: 143 cal/ 100g
Next, compare the daily consumption with the recommended value and print if the child is undernourished or not.

 

Data for BMI Calculation:
                                                             BMI = weight/(height**2)*703

 

         Weight in pounds and height in inches 

 



```python
# To store the data in list
def details(name , age , gender , height , weight, s,c):
    l = [name , age , gender , height , weight, s ,c]
    return l

# Calculate the BMI 
def BMI(height , weight):
    Body_Mass_Index = weight/(height**2)*703
    t = " "
    if(Body_Mass_Index < 16):
        t = "Severely Underweight"
    
    elif(Body_Mass_Index >= 16 and Body_Mass_Index < 18.5):
        t = "UnderWeight"
        
    elif(Body_Mass_Index >= 18.5 and Body_Mass_Index < 25):
        t = "Healthy"

    elif(Body_Mass_Index >= 25 and Body_Mass_Index < 30 ):
        t = "Overweight"
        
    else:
        t = "Obese"
        
    return t

# Recommended Value Calories Requirement
def calories_requirement (age):
    minimum_daily_calorie_requirement = 0
    if(age > 0 and age <= 2 ):
        minimum_daily_calorie_requirement = 800
    elif(age > 2 and age <= 4):
        minimum_daily_calorie_requirement = 1400
    elif(age > 4 and age <=8):
        minimum_daily_calorie_requirement = 1800
        
    return minimum_daily_calorie_requirement
        
# Daily Consumption of Calories
def Daily_Consumption():
    sum = 0
    Milk = input("Milk : YES or NO ")
    Milk = Milk.lower()
    if(Milk == "yes"):
        quantity = float(input("Enter Quantity of MILK in grams: "))
        sum+=quantity
    Egg = input("Egg : YES or NO ")
    Egg = Egg.lower()
    if(Egg == "yes"):
        quantity = float(input("Enter Quantity of Egg in grams: "))
        sum+=quantity* 1.55
    Rice = input("Rice : YES or NO ")
    Rice = Rice.lower()
    if(Rice == "yes"):
        quantity = float(input("Enter Quantity of Rice in grams: "))
        sum+=quantity*1.30
    Lentils = input("Lentils : YES or NO ")
    Lentils = Lentils.lower()
    if(Lentils == "yes"):
        quantity = float(input("Enter Quantity of Lentils in grams: "))
        sum+=quantity*1.13
    Vegetable = input("Vegetable : YES or NO ")
    Vegetable = Vegetable.lower()
    if(Vegetable == "yes"):
        quantity = float(input("Enter Quantity of Vegetable in grams: "))
        sum+=quantity*0.85
    Meat = input("Meat : YES or NO ")
    Meat = Meat.lower()
    if(Meat == "yes"):
        quantity = float(input("Enter Quantity of Meat in grams: "))
        sum+=quantity*1.43
        
    return sum

# No.of Children
n = int (input("Enter No. Of Children "))
data = []
for i in range(0,n):
    print((i+1),"Child data")
    name = input("Enter the name ") 
    age = int(input("Enter the age of "))
    gender = input ("Enter the gender F/M of ")
    height = float(input("Enter the height in inches "))
    weight = float(input("Enter the weight in pounds "))
    s = BMI(height , weight)
    a = Daily_Consumption()
    b = calories_requirement (age)
    if(a < b):
        c = "Undernourished"
    else:
        c = "Not Undernourished"

    data.append(details(name , age , gender , height , weight , s , c))
print (data)
```
